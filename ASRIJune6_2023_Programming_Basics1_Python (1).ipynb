#Fibonacci Sequence
I'm going to look at the Fibonacci Sequence. In the Fibonacci sequence each term the sum of the two that preceed it. The sequence is the series of numbers that result from the application of the rule.

In this example, I’ll introduced a *while loop*. This loop will executute while the condition is true, and it will terminate when it is no longer true. This is very useful if you don’t know how many loops it will take to calculate soemthing! There are also loops with counters that enable a loop to be executed a particular number of times.

**Note that the loop code is indented.** Blocks of code are identified only by indentation.
Q: How would I need to change the condition to print 1 more term? (The code is from the official Python Tutorial.)


```python
a, b=0, 1
#These are the first two terms of the Fibonacci series, each subsequent term is the sum of the two that went before.
# The numbers that are require to
while a<10:
      print(a)
      a, b=b,a+b
```

##Logical Flow of the Code

Let’s see what the while loop does, we start from a=0, b=1 and print a

a=0, b=1, print 0

a=1, b=1, print 1

a=1, b=2, print 1

a=2, b=3, print 2

a=3, b=5, print 3

a=5, b=8, print 5

a=8, b=13, print 8

a=13, b=21, print nothing


The keyword argument end can be used to avoid the newline after the output, I'll used it to replace the newline with a "," see below

```python
a, b=0, 1
while a<10:
      print(a, end=',')
      a, b=b,a+b
```
